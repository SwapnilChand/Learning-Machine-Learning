VISUALIZATION IN MATPLOTLIB AND SEABORN

- age, name and potential/release clause comparision
- age, name and nationality comparision

```python
age = fifa['Age'][0:5].sort_values()
name = fifa['Name'][0:5].sort_values()
nation = fifa['Nationality'][0:5].sort_values()
clause = fifa['Release Clause'][0:5].sort_values()
potential = fifa['Potential'][0:5].sort_values()

# plt.plot(color ='green',linewidth=3, marker='o',
        #  markersize=15, linestyle='--')
```

using axes() for making multiple plots



```python
fig = plt.figure(figsize = (5, 4))

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax2 = fig.add_axes([1, 0.1, 0.8, 0.8])

plt.xlim(0, 25)
plt.ylim(0, 25)

plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')

ax1.plot(name, potential)
ax2.plot(name, age)

```

```python
fig = plt.figure(figsize= (11, 6))

ax = fig.add_axes([1,1,1,1])
ax3 = ax.plot(name, age)
ax4 = ax.plot(nation, clause)

ax.set_title("Linear Graph")
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")

ax.legend(labels = ('name against age', 'name against clause'))

```
